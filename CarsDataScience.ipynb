Categorical vs Categorical

```python
popular_brands = ad_sub["standard_make"].value_counts().nlargest(10).index
```

```python
popular_brands
```

```python
most_common = ad_sub[ad_sub["standard_make"].isin(popular_brands)]
```

```python
most_common.groupby(["standard_make", "fuel_type"])["price"].mean().unstack("fuel_type").plot.bar()
```

```python
brand_bodytype = most_common.groupby(["standard_make","body_type"])["price"].mean().unstack()
sns.heatmap(data = brand_bodytype)
```
